```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN',
'londON_StockhOlm',

'Budapest_PaRis', 'Brussels_londOn'],
'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],
'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],
'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )',
'12. Air France', '"Swiss Air"']})
```

```python
df
```

```python
##Function to get To City
def split_To(city):
    from_to =city.split('_')
    to=from_to[1]
    return to
```

```python
##Function to get From City
def split_From(city):
    from_to =city.split('_')
    from_city=from_to[0]
    return from_city
```

```python
split_From('LoNDon_paris')
```

```python
df['From']=df['From_To'].apply(split_From)
```

```python
df
```

```python
df['To'] = df['From_To'].apply(split_To)
```

```python
df
```

```python
df.drop('From_To',axis=1,inplace=True)
```

```python
df
```

```python
##Changing the City details to proper format
df['To'] = df['To'].apply(lambda x:x.title())
df['From'] = df['From'].apply(lambda x:x.title())
```

```python
df
```

```python
##Null number indices
null_num_index = df[df['FlightNumber'].isnull()].index
```

```python
##Function to fill null index values
def fillNum(df):
    null_num_index = df[df['FlightNumber'].isnull()].index
    for num in null_num_index:
        df['FlightNumber'].iloc[num] = df['FlightNumber'].iloc[num-1]+10.0
    return df
```

```python
df=fillNum(df)
```

```python
df
```

```python
##Splitting the recent delays
df[['Delay1','Delay2','Delay3']] = pd.DataFrame(df['RecentDelays'].to_list(),index=df.index)
```
